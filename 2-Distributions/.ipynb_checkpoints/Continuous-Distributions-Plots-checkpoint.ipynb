## Distribuição Gamma $\Gamma$

$$
\Gamma(x) = \int_{x=0}^{\infty} t^{x-1} e^{-t} dt
$$

Onde $\mathcal{X}$, imagem de $X$ é o conjunto $[0,\infty)$

```python
# Define x range for the gamma function
x = np.linspace(0.1, 5, 400)
gamma_values = gamma(x)
```

A função de densidade de probabilidade $\text{pdf}$ de $X$ é expressa por:

$$
f_X(x|\alpha,\beta) = \frac{1}{\Gamma(\alpha) \beta^{\alpha}} x^{\alpha-1} e^{-x/\beta}
$$

Onde:
* $0<x<\infty$
* $\alpha>0$
* $\beta>0$

Casos particulares notáveis:
* $\alpha = 1$<br>
Distribuição exponencial $f_X(x|\beta) = \frac{e^{-x/\beta}}{\beta}$

```python
x_dist = np.linspace(0, 20, 400)

f,axs=plt.subplots(1,3,figsize=(15,3))
axs[0].plot(x, gamma_values, label=r'$\Gamma(x)$', color='blue')
axs[0].set_title(r'$\Gamma(x)$')
axs[0].set_xlabel('$x$')
axs[0].set_xlim(0,)

b = [1,2,5,10]
cores_pdf = plt.cm.Blues(np.linspace(.2,.8,len(b)))
cores_cdf = plt.cm.Reds(np.linspace(.2,.8,len(b)))

for i,b in enumerate(zip(b)):
    pdf_values = gamma_dist.pdf(x_dist,a=1, scale=b)
    cdf_values = gamma_dist.cdf(x_dist,a=1, scale=b)    
    axs[1].plot(x_dist, pdf_values, color=cores_pdf[i], label=f'b={b}')
    axs[2].plot(x_dist, cdf_values, color=cores_cdf[i], label=f'b={b}')

# for i,(a,b) in enumerate(zip(alfa, beta)):
#     pdf_values = gamma_dist.pdf(x_dist, a=a, scale=b)
#     cdf_values = gamma_dist.cdf(x_dist, a=a, scale=b)    
#     axs[1].plot(x_dist, pdf_values, color=cores_pdf[i], label=fr'$\alpha=${a}, $\beta=${b}')
#     axs[2].plot(x_dist, cdf_values, color=cores_cdf[i], label=fr'$\alpha=${a}, $\beta=${b}')
axs[1].legend()
axs[2].legend()



# for ax,t in zip((axs[1],axs[2]),['PDF','CDF']):
#     ax.set_title(t)
#     ax.set_xlabel('$x$')
#     ax.set_xlim(0,)
```

```python
alpha_values = [0.5,1,50,100]
beta_values  = [x/3 for x in alpha_values]

# Geração de valores de x no intervalo [0,1]
x = np.linspace(0, 1, 1000)

# Plotando diferentes curvas da distribuição Beta
plt.figure(figsize=(8,6))
for alpha, beta_param in zip(alpha_values, beta_values):
    y = beta.pdf(x, alpha, beta_param)
    plt.plot(x, y, label=fr'$\alpha =${alpha}, $\beta=${beta_param}')

# Configurações do gráfico
plt.title('Distribuição Beta para diferentes parâmetros')
# plt.ylim(0,100)
plt.xlabel('x')
plt.ylabel('Densidade de Probabilidade')
plt.legend()
plt.grid()
plt.show()
```
