# Variáveis aleatórias multiplas

## Importações e configurações de ambiente

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
from mpl_toolkits.mplot3d import Axes3D
from scipy.stats import gaussian_kde
import warnings
```

```python
# Melhora qualidade dos plots no jupyter
%config InlineBackend.figure_format = 'retina'

# Ignora avisos
warnings.filterwarnings("ignore")
```

## Distribuição conjunta (*joint*) e marginal

Parâmetro para geração dos dados sintéticos de altura e peso

```python
media_altura = 170
desvio_altura = 10
media_peso  = 70
desvio_peso = 15
correlacao = 0.7
```

```python
# Matriz de covariância
covariancia = [[desvio_altura**2, correlacao * desvio_altura * desvio_peso],
                     [correlacao * desvio_altura * desvio_peso, desvio_peso**2]]
```

```python
N = 1000
alturas, pesos = np.random.multivariate_normal([media_altura, media_peso], covariancia, N).T
```

```python
# Cria dataframe
data = pd.DataFrame({'altura (cm)': alturas,'peso (kg)': pesos})
```

```python
def pdf(dados):
    kde = gaussian_kde(dados)
    x = np.linspace(dados.min(),dados.max(),dados.shape[0])
    y = kde(x)
    return x,y
```

```python
# Valores adequados para a pdf dos pesos e alturas
x_pesos, pdf_pesos = pdf(data['peso (kg)'])
x_alturas, pdf_alturas = pdf(data['altura (cm)'])
```

Cálculo da probabilidade conjunta $f_{XY}(x,y)$

```python
# Gerando os valores para a superfície 3D
x = np.linspace(min(alturas), max(alturas), 100)
y = np.linspace(min(pesos), max(pesos), 100)
x_grid, y_grid = np.meshgrid(x, y)

# Calculando a densidade de probabilidade para a grade de pontos
kde_2d = gaussian_kde([alturas, pesos])
z_grid = kde_2d(np.vstack([x_grid.ravel(), y_grid.ravel()])).reshape(x_grid.shape)
```

```python
f=plt.figure(figsize=(14, 5))
# f=plt.figure()
gs = GridSpec(1, 3, width_ratios=[2, 2, 3])  # Aumentando a proporção do gráfico 3D

# PDF dos pesos
ax1 = f.add_subplot(gs[0])
ax1.plot(x_pesos, pdf_pesos,label='Pesos',color='tab:blue')
ax1.set_xlabel('Pesos (kg)')
ax1.grid(alpha=0.3)


# # PDF das alturas
ax_sec = ax1.twiny()
ax_sec.plot(x_alturas, pdf_alturas, label='Alturas',color='tab:red')
ax_sec.set_xlabel('Altura (cm)')
ax_sec.set_ylim(0,)

for ax in (ax1,ax_sec):
    ax.minorticks_on()

# Dispersão entre a altura e peso
ax2 = f.add_subplot(gs[1])
ax2.scatter(data['peso (kg)'], data['altura (cm)'],alpha=0.3)
ax2.set_xlabel('Pesos (kg)')
ax2.set_ylabel('Altura (cm)')
ax2.grid(alpha=0.3)

ax3 = f.add_subplot(gs[2],projection='3d')
# ax3.set_box_aspect([1,.5,2])
ax3.plot_surface(x_grid, y_grid, z_grid, cmap='plasma', edgecolor='w', alpha=0.8, lw=.5)
ax3.xaxis.pane.fill = False   # Remove color on the XZ plane
ax3.yaxis.pane.fill = False   # Remove color on the YZ plane
ax3.zaxis.pane.fill = False   # Remove color on the XY plane
ax3.axis('off')
# plt.tight_layout()
plt.show()
```
